**Problem Statement:**

Suppose a house's value in a specific area decreases by half each year, meaning that a house worth 1 million dollars today will be worth 0.5 million in one year, 0.25 million in two years, and so forth. This phenomenon is referred to as exponential decay. Given the current value of a house, our goal is to create a plot depicting its projected worth over the course of 0 to 10 years.


````python
import math

def future_worth(t):
    '''
    Returns worth after t years
    At t = 0, worth = 700,000
    '''
    base_worth = 700000
    return base_worth * math.pow(2, -t)
````

````python
# Expected output: a plot of the price of the house over time
import matplotlib.pyplot as plt
import numpy as np

# REPLACE _ WITH APPROPRIATE QUANTITY
# hint: np.arange(n) is equivalent to [0,1,2,..,n-1]
x = np.arange(11)

y = []
for t in x:
    # REPLACE _ WITH APPROPRIATE QUANTITY
    y.append(future_worth(t))

# Create the plot
plt.plot(x,y)

# Sets the title of the plot
plt.title('House worth vs time')

# Displays the plot
plt.show()
````

###Most Commonly Used ML Models
#### 1. Linear Regression
Linear Regression is used to predict the value of a dependent variable based on one or more independent variables. It assumes a linear relationship between these variables. It is best used for trend analysis and forecasting in finance, sales, marketing, and other fields.

#### 2. Logistic Regression
Logistic Regression is used for binary classification problems, where the output is categorical and represents two classes. It estimates the probability that a given instance belongs to a particular category. It’s widely used in medicine (disease prediction), finance (credit scoring), and more.

#### 3. Decision Trees
Decision Trees are used for both classification and regression problems. They model decisions and their possible consequences, representing an upside-down tree structure. It's a good choice when interpretability is important, like in strategic planning or medical diagnosis.

#### 4. Random Forest
Random Forest is an ensemble method using multiple decision trees for classification, regression, and other tasks. It improves the predictive accuracy and controls over-fitting. It’s versatile and can be used in various domains, including e-commerce, finance, and healthcare.

#### 5. Support Vector Machines (SVM)
Support Vector Machines are used for classification and regression tasks, working well for both linear and non-linear relationships. They are effective in high-dimensional spaces, making them suitable for text classification, image recognition, and more.

#### 6. K-Nearest Neighbors (KNN)
K-Nearest Neighbors is a simple, instance-based learning algorithm used for classification and regression. It predicts the output based on the majority vote or average of the K nearest instances. It is best suited for small datasets where the relationship between variables is complex.

#### 7. Neural Networks
Neural Networks are a set of algorithms modeled after the human brain, designed to recognize patterns. They are highly flexible and capable of learning complex relationships, making them suitable for image recognition, speech processing, and more.


#### 8. Naive Bayes
Naive Bayes is a probabilistic model based on applying Bayes' theorem with strong independence assumptions between the features. It is particularly used for text classification, including spam filtering and sentiment analysis.

###Understanding Neural Networks

A neural network is a type of computer program or model that is inspired by the structure and functioning of the human brain. It's used to solve a wide range of problems, especially those that involve recognizing patterns, making decisions, or finding relationships in data.


A neural network system consists of
1. **Basic Building Block - The Neuron:** In a neural network, the basic building block is called a "neuron" or "node." Neurons are like little decision-makers. They take some input, process it, and produce an output.

2. **Layers of Neurons:** Neurons are organized into "layers." You can think of these layers like different groups of experts, each looking at a specific aspect of the problem. The input data goes into the first layer, and the output comes from the last layer.

3. **Connections and Weights:** Neurons in one layer are connected to neurons in the next layer. Each connection has a "weight." Weights are like the strength of influence between neurons.

4. **Learning from Data:** The neural network learns from data. It adjusts the weights on connections by looking at many examples. It learns how to make better decisions based on the data it sees. This process is called "training" the neural network.

5. **Making Predictions:** Once a neural network is trained, you can use it to make predictions.


###Understanding our Dataset:

####[CIFAR-10 Dataset](https://www.kaggle.com/c/cifar-10/)
1. It comprises 60,000 32x32 color images, categorized into 10 different classes (e.g., airplane, cat, automobile), with 6,000 images per class.

2. It is split into a training set of 50,000 images and a test set of 10,000 images. Each image is labeled with a class identifier.
3. CIFAR-10 is commonly used for image recognition tasks, serving as a benchmark for machine learning and computer vision algorithms.

````python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()

# Class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Display basic information about the dataset
print("Training data shape:", train_images.shape)
print("Number of training samples:", len(train_images))
print("Test data shape:", test_images.shape)
print("Number of test samples:", len(test_images))
print("Number of classes:", len(class_names))

# Function to visualize images
def plot_images(images, labels, predictions=None):
    assert len(images) == len(labels) == 9

    fig, axes = plt.subplots(3, 3, figsize=(10, 10))
    axes = axes.ravel()

    for i in np.arange(0, 9):
        axes[i].imshow(images[i])
        axes[i].set_title("True: %s\n" % class_names[int(labels[i])], fontsize=12)
        axes[i].axis('off')
        plt.subplots_adjust(wspace=1)

# Show 9 random images from the training set
indices = np.random.randint(0, train_images.shape[0], size=9)
images = train_images[indices]
labels = train_labels[indices]
plot_images(images, labels)
plt.show()
````

###Understanding Activation Functions
####ReLU (Rectified Linear Unit):
**What It Does:** If a number is positive, ReLU keeps it as is; if the number is negative, ReLU turns it into a zero.

**Why It's Good:** It speeds up training and helps with big, deep networks. It's also really simple and quick to compute.

**When to Use It:** Use it in hidden layers in a neural network most of the time.

####Softmax:
**What It Does:** It turns scores into probabilities.

**Why It's Good:** It's great for multi-class classification problems where each input could belong to multiple categories.

**When to Use It:** Use it in the output layer of a neural network when you're predicting probabilities of different classes.

###Building our Model

````python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt


#Add TODO

# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

# Convert testing and training data into one-hot encoding
#Fill in _ with the appropriate quantity
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# Define the model
#Give the entire model sequence
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
#Fill in _ with the appropriate quantity. We fit the model to training images and training labels.
history = model.fit(train_images, train_labels, epochs=10, validation_split=0.2)

# Evaluate the model - Models are evaluated with our test images, test labels
#Fill in _ with the appropriate quantity
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('\nTest accuracy:', test_acc)

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
````

Now, we will take a random image from the test dataset, and make a prediction on it.

````python
import numpy as np
import matplotlib.pyplot as plt

# Class names for CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Pick a random image from the test dataset
random_index = np.random.randint(0, len(test_images))
random_image = test_images[random_index]
random_image_to_show = random_image * 255  # Rescale back for displaying

# Predict
predictions = model.predict(np.array([random_image]))
predicted_label = np.argmax(predictions[0])

# Show the image
plt.figure()
plt.imshow(random_image_to_show.astype('uint8'))
plt.title(f'Predicted Label: {class_names[predicted_label]}')
plt.show()
````
